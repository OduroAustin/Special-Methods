# Python OOP: Special (Dunder) Methods

Special methods are predefined methods in Python that start and end with double underscores (e.g., `__init__`, `__str__`). Often called **dunder methods**, they allow you to hook into Python's built-in behavior and implement operator overloading, object initialization, and custom string representations.

Instead of calling these methods directly, Python triggers them behind the scenes when specific syntax is used.

## Core Special Methods

*   `__init__`: Initializes a newly created object.
*   `__str__`: Defines the user-friendly string representation triggered by `print()`.
*   `__add__`: Overloads the addition (`+`) operator.

---

## Code Example

```python
class Book:
    def __init__(self, title: str, pages: int):
        self.title = title
        self.pages = pages

    def __str__(self) -> str:
        return f"'{self.title}' ({self.pages} pages)"

    def __add__(self, other: "Book") -> int:
        return self.pages + other.pages


if __name__ == "__main__":
    book1 = Book("Python Basics", 200)
    book2 = Book("Advanced Python", 350)

    print(book1)
    print(book2)

    total_pages = book1 + book2
    print(f"Total reading material: {total_pages} pages")
